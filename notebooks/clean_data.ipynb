```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read dataset
houses = pd.read_csv('/Volumes/Untitled/MDDB/02-03 Artificial Intelligence for Business/Raw data/housing_data.csv')
```

```python
        #read dataset
        #houses = pd.read_csv('/Volumes/Untitled/MDDB/02-03 Artificial Intelligence for Business/Raw data/housing_data.csv')
        
        # replace nan
        houses['aantalBadkamers'] = houses['aantalBadkamers'].replace(np.NaN, houses["aantalBadkamers"].mean())
        houses['koopPrijs'] = houses['koopPrijs'].replace(np.NaN, houses["koopPrijs"].mean())

        #remove outlier oppervlakte (1)
        meansurface = houses['oppervlakte'].mean()
        houses['oppervlakte'].replace(1, meansurface, inplace=True)

        # remove part of the output
        #houses['bouwjaar'] = [i[:4] for i in houses['bouwjaar']]
        houses['bouwjaar'] = houses['bouwjaar'].str[-4:]
        houses['postcode_numbers'] = [i[:4] for i in houses['postcode']] #store in new column
        
        # categorieObject 0 = woonhuis, 1 = appartement
        houses['categorieObject'].replace('<{Woonhuis}>', 0, inplace=True)
        houses['categorieObject'].replace('<{Appartement}>', 1, inplace=True)

        # encode energy labels
        houses['energielabelKlasse'] = houses['energielabelKlasse'].fillna(0)
        houses['energielabelKlasse'].replace('A+++++', 1, inplace=True)
        houses['energielabelKlasse'].replace('A++++', 2, inplace=True)
        houses['energielabelKlasse'].replace('A+++', 3, inplace=True)
        houses['energielabelKlasse'].replace('A++', 4, inplace=True)
        houses['energielabelKlasse'].replace('A+', 5, inplace=True)
        houses['energielabelKlasse'].replace('A', 6, inplace=True)
        houses['energielabelKlasse'].replace('B', 7, inplace=True)
        houses['energielabelKlasse'].replace('C', 8, inplace=True)
        houses['energielabelKlasse'].replace('D', 9, inplace=True)
        houses['energielabelKlasse'].replace('E', 10, inplace=True)
        houses['energielabelKlasse'].replace('F', 11, inplace=True)
        houses['energielabelKlasse'].replace('G', 12, inplace=True)

        # split soortWoning and encode soortwoning1
        houses[['woningtype1','woningtype2']] = houses.soortWoning.str.split("}>",1,expand=True)
        houses['woningtype1'] = houses['woningtype1'].str[2:]

        houses['woningtype1'].replace('woonboot', 0, inplace=True)
        houses['woningtype1'].replace('eengezinswoning', 1, inplace=True)
        houses['woningtype1'].replace('tussenverdieping', 2, inplace=True)
        houses['woningtype1'].replace('woonboerderij', 3, inplace=True)
        houses['woningtype1'].replace('portiekflat', 4, inplace=True)
        houses['woningtype1'].replace('galerijflat', 5, inplace=True)
        houses['woningtype1'].replace('villa', 6, inplace=True)
        houses['woningtype1'].replace('beneden + bovenwoning', 7, inplace=True)
        houses['woningtype1'].replace('bovenwoning', 8, inplace=True)
        houses['woningtype1'].replace('benedenwoning', 9, inplace=True)
        houses['woningtype1'].replace('maisonnette', 10, inplace=True)
        houses['woningtype1'].replace('herenhuis', 11, inplace=True)
        houses['woningtype1'].replace('penthouse', 12, inplace=True)
        houses['woningtype1'].replace('bungalow', 13, inplace=True)
        houses['woningtype1'].replace('landhuis', 14, inplace=True)
        houses['woningtype1'].replace('grachtenpand', 15, inplace=True)
        houses['woningtype1'].replace('dubbel benedenhuis', 16, inplace=True)
        houses['woningtype1'].replace('landgoed', 17, inplace=True)
        houses['woningtype1'].replace('studentenkamer', 18, inplace=True)
        houses['woningtype1'].replace('stacaravan', 19, inplace=True)
        houses['woningtype1'].replace('portiekwoning', 20, inplace=True)

        # change dtype
        houses['publicatieDatum'] = pd.to_datetime(houses['publicatieDatum'])
        houses['datum_ondertekening'] = pd.to_datetime(houses['datum_ondertekening'])
        
        #Round # of bathrooms
        houses['aantalBadkamers']=houses['aantalBadkamers'].round(decimals=0)
        
        # set threshold house price 50000
        houses = houses.drop(houses[houses['koopPrijs']<=50000].index)
```

```python
del houses['globalId.1'] #redundant
del houses['soortWoning'] # this one is encoded --> redundant
del houses['perceelOppervlakte']
```

```python
#check missing values
num_missing = houses.isnull().sum()
percent = num_missing / houses.isnull().count()

df_missing = pd.concat([num_missing, percent], axis=1, keys=['MissingValues', 'Fraction'])
df_missing = df_missing.sort_values('Fraction', ascending=False)
df_missing[df_missing['MissingValues'] > 0] 
```
